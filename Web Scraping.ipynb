```python
#Scrapping halaman web
import requests
from bs4 import BeautifulSoup
import json

res = requests.get('https://www.bankmega.com/id/promo/')
soup = BeautifulSoup(res.content, "html.parser")

script = soup.find_all('script')[10].text.strip()[42:-7865]
print(script)
```

```python
script1 = script.replace("\t","")
script2 = script1.replace("\n","")
script2
```

Mengambil bagian informasi yang dipisahkan tanda koma


```python
#mencari lokasi tanda koma
koma = []
for i in range(len(script2)) :
  if script2[i] == ",":
    koma.append(i)
print(koma)
```

```python
len(koma)
```

```python
#memisahkan dan menyimpan bagian informasi
data1 = []
data1.append(script2[0:koma[0]]) #untuk menyimpan data informasi pertama, karena pada bagian ini tidak diawali tanda koma
i = 1

for k in range(len(koma)-1):
  bagian_info = script2[koma[i-1]:koma[i]]
  i += 1
  data1.append(bagian_info)
data1.append(script2[koma[-1]:len(script2)])
print(data1)
```

```python
len(data1)
```

```python
data1[0:10]
```

Simpan setiap bagian informasi ke dalam list yang berbeda, misal informasi judul simpan ke dalam list judul


```python
#cari informasi yang mengandung kata "title" kemudian simpan ke list judul
judul = []
for k in data1 :
  if "title" in k:
    judul.append(k)
print(judul)
```

```python
#untuk yang informasi slug
slug = []
for k in data1 :
  if "slug" in k:
    slug.append(k)
print(slug)
```

```python
#untuk informasi tanggal dan bulan expired
expired = []
for k in data1 :
  if "expired_date" in k:
    expired.append(k)
print(expired)
```

```python
#untuk informasi tahun expired
tahun = []
for k in data1 :
  if len(k) < 7 and "20" in k:
    tahun.append(k)
print(tahun)
```

```python
jam_akhir = []
for k in data1 :
  if "malam" in k or "a.m" in k or "p.m" in k:
    jam_akhir.append(k)
print(jam_akhir)
```

```python
gambar = []
for k in data1 :
  if "featured_image" in k :
    gambar.append(k)
print(gambar)
```

```python
kategori = []
for k in data1 :
  if "categories" in k :
    kategori.append(k)
print(kategori)
```

```python
kota = []
for k in data1 :
  if "city" in k or "cities" in k:
    kota.append(k)
print(kota)
```

```python
print("banyak judul : ", len(judul))
print("banyak slug : ", len(slug))
print("banyak expired : ", len(expired))
print("banyak tahun : ", len(tahun))
print("banyak jam_akhir : ", len(jam_akhir))
print("banyak kategori : ", len(kategori))
print("banyak gambar : ", len(gambar))
print("banyak kota : ", len(kota))
```

Dikarenakan ada informasi yang mempunyai beberapa kota, maka kota-kota tersebut akan dibuat terpisah dan informasi lainnya yang terkait akan diduplikat

```python
#mencari lokasi data yang mengandung city tanpa cities
kota1 = []
i = 0
for a in kota:
  if "cities" not in a:
    kota1.append(i)
  i += 1
print(kota1)
```

```python
#menduplikat informasi lainnya untuk kota yang hanya tertulis city
for i in range(len(kota1)):
  isi = kota1[i]
  judul.insert(isi,judul[kota1[i]-1])

for i in range(len(kota1)):
  isi = kota1[i]
  slug.insert(isi,slug[kota1[i]-1])

for i in range(len(kota1)):
  isi = kota1[i]
  expired.insert(isi,expired[kota1[i]-1])

for i in range(len(kota1)):
  isi = kota1[i]
  tahun.insert(isi,tahun[kota1[i]-1])

for i in range(len(kota1)):
  isi = kota1[i]
  jam_akhir.insert(isi,jam_akhir[kota1[i]-1])

for i in range(len(kota1)):
  isi = kota1[i]
  gambar.insert(isi,gambar[kota1[i]-1])

for i in range(len(kota1)):
  isi = kota1[i]
  kategori.insert(isi,kategori[kota1[i]-1])

```

Memperbaiki tampilan data dengan menghilangkan kata atau tanda baca yang tidak digunakan

```python
judul_bersih = []
for a in judul:
  judul_baru1 = a.replace("{title: ","") 
  judul_baru2 = judul_baru1.replace("\"","")  
  judul_baru3 = judul_baru2.replace(",","")
  judul_bersih.append(judul_baru3)
print(judul_bersih) 
```

```python
slug_bersih = [] 
for a in slug:
  slug1 = a.replace(",slug: ","") 
  slug2 = slug1.replace("\"","")
  slug_bersih.append(slug2)
print(slug_bersih)
```

```python
# untuk tanggal expired dan tahun sebelum diperbaiki disatukan terlebih dahulu
i = 0
for a in tahun:
  expired.insert(i, expired[i]+a)
  i += 1
expired[0:5]
```

```python
#setelah berhasil digabungkan, baru diperbaiki tampilannya
waktu_bersih = []
for a in expired:
  expired1 = a.replace(",expired_date:","")
  expired2 = expired1.replace(",","")
  expired3 = expired2.replace("\"","")
  waktu_bersih.append(expired3)
print(waktu_bersih)
```

```python
jam_akhir_bersih = []
for a in jam_akhir:
  jam_akhir1 = a.replace(",","")
  jam_akhir2 = jam_akhir1.replace("\"","")
  jam_akhir_bersih.append(jam_akhir2)
print(jam_akhir_bersih)
```

```python
gambar_bersih = []
for a in gambar:
  gambar1 = a.replace(",featured_image:","")
  gambar2 = gambar1.replace("\"","")
  gambar_bersih.append(gambar2)
print(gambar_bersih)
```

```python
kategori[0:5]
```

```python
kategori_bersih = []
for a in kategori:
  kategori1 = a.replace(",categories:[{category:","")
  kategori2 = kategori1.replace("}","")
  kategori3 = kategori2.replace("\"","")
  kategori_bersih.append(kategori3)
print(kategori_bersih)
```

```python
kota_bersih = []
for a in kota:
  kota1 = a.replace(",cities:[{city:","")
  kota2 = kota1.replace(",{city:","")
  kota3 = kota2.replace("\"","")
  kota4 = kota3.replace("}","")
  kota_bersih.append(kota4)
print(kota_bersih)
```

Menggabungkan List-list Informasi Menjadi Satu Data Frame

```python
import pandas as pd

# Calling DataFrame constructor on list
df = pd.DataFrame(list(zip(judul_bersih, slug_bersih, waktu_bersih, jam_akhir_bersih, gambar_bersih, kategori_bersih, kota_bersih)), columns =['Judul', 'Slug', 'Tanggal Expired', 'Jam Berakhir', 'Gambar', 'Kategori', 'Kota'])
df
```
